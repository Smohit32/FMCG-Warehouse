Just to show the Avgno_workers in line chart , i have groupby the warehouses into 50 , so that we can have clean and clear line chart.

```python
average_workers_per_warehouse = x['workers_num'].mean()

# Group the data by groups of 50 warehouses
data_sorted = x.sort_values(by='Ware_house_ID')
data_sorted.reset_index(drop=True, inplace=True)

# Define the number of warehouses per group
group_size = 50

# Create groups and calculate the average number of workers per group
grouped_data = data_sorted.groupby(data_sorted.index // group_size)['workers_num'].mean()

# Plot the average number of workers per group of 50 warehouses
plt.figure(figsize=(12, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='b')
plt.title('Average Number of Workers per Group of 50 Warehouses')
plt.xlabel('Group (each group contains 50 warehouses)')
plt.ylabel('Average Number of Workers')
plt.grid(True)
plt.show()
```

Q11. What is the correlation between the number of workers and the number of reported storage issues in the last 3 months?

```python
workers_and_issues =x[['workers_num','storage_issue_reported_l3m']]
Correalation = workers_and_issues.corr().loc['workers_num','storage_issue_reported_l3m']
print(Correalation)
plt.figure(figsize=(8, 6))
plt.scatter(x['workers_num'], x['storage_issue_reported_l3m'], color='skyblue', alpha=0.7)
plt.title('Correlation Between Number of Workers and Reported Storage Issues')
plt.xlabel('Number of Workers')
plt.ylabel('Reported Storage Issues')

plt.text(0.5, 0.95, f'Correlation coefficient: {Correalation:.2f}', ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

ANS. Our analysis indicates that there is a negative correlation between the number of workers and the number of reported storage issues over the last three months. This negative correlation suggests that as the number of workers increases, the number of reported storage issues tends to decrease. In other words, warehouses with more staff are experiencing fewer storage problems.

Q20. Investigate the relationship between warehouse establishment year and the number of breakdowns reported in the last 3 months.


```python
Correlation_establishment_breakdowns = x[['wh_est_year','wh_breakdown_l3m']].corr().loc['wh_est_year','wh_breakdown_l3m']
print(Correlation_establishment_breakdowns)

plt.figure(figsize=(8, 5))
plt.scatter(x['wh_est_year'], x['wh_breakdown_l3m'], color='skyblue', alpha=0.7)
plt.title('Correlation Between Warehouse Establishment Year and Breakdowns')
plt.xlabel('Warehouse Establishment Year')
plt.ylabel('Warehouse Breakdowns in Last 3 Months')

plt.text(0.5, 0.95, f'Correlation coefficient: {Correlation_establishment_breakdowns:.2f}',
         ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()
```

The relationship between warehouse establishment year and the number of breakdowns reported in the last 3 months shows a moderate negative correlation of approximately -0.3988. This suggests that newer warehouses (established more recently) tend to experience fewer breakdowns compared to older warehouses.
